## Coffee Prediction Prediction

Given *data about coffee*, let's try to predict the **average production** of coffee in a given country.

We will use a random forest regression model to make our predictions.

Data source: https://www.kaggle.com/datasets/yamaerenay/ico-coffee-dataset-worldwide

### Getting Started

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestRegressor
```

```python
df_paths = [
    'archive/domestic-consumption.csv',
    'archive/exports-calendar-year.csv',
    'archive/exports-crop-year.csv',
    'archive/gross-opening-stocks.csv',
    'archive/total-production.csv'
]
```

```python
dfs = [pd.read_csv(df_path) for df_path in df_paths]
```

```python
dfs[0].mean(axis=1, numeric_only=True)
```

```python
def get_means(df):
    df = df.copy()
    countries = df[df.columns[0]]
    means = df.mean(axis=1, numeric_only=True)
    df = pd.concat([countries, means], axis=1)
    df.columns = ['country', countries.name]
    return df
```

```python
get_means(dfs[0]).merge(get_means(dfs[1]), on='country')
```

```python
def make_df(dfs):
    # Process all dataframes into one dataframe
    processed_dfs = []

    for df in dfs:
        processed_dfs.append(get_means(df))

    # Merge DataFrames 
    df = processed_dfs[0]

    for i in range(1, len(processed_dfs)):
        df = df.merge(processed_dfs[i], on='country')

    return df
```

```python
data = make_df(dfs)
```

```python
data
```

```python
df = data.copy()
```

```python
df.info()
```

```python
# Drop country column
df = df.drop('country', axis=1)
```

```python
# Split df into X and y
y = df['total_production']
X = df.drop('total_production', axis=1)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
# Scale X
scaler = StandardScaler()
scaler.fit(X_train)

X_train = pd.DataFrame(scaler.transform(X_train), index=X_train.index, columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), index=X_test.index, columns=X_test.columns)
```

```python
X_train
```

### Training/Results

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
print("Model Trained.")
```

```python
y_pred = model.predict(X_test)

rmse = np.sqrt(np.mean((y_test - y_pred)**2))
print("RMSE: {:.2f}".format(rmse))

r2 = 1 - (np.sum((y_test - y_pred)**2) / (np.sum((y_test - y_test.mean())**2)))
print("R2 Score: {:.5f}".format(r2))
```

```python
model.score(X_test, y_test)
```
