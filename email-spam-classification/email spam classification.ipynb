## ðŸ“§ Email Spam Classification

Given *data about emails*, let's try to predict whether a given email will be **spam** or not.

We will use a Tensorflow/Keras neural network with word embeddings to make our predictions.

Data source: https://www.kaggle.com/datasets/chandramoulinaidu/spam-classification-for-basic-nlp

### Importing Libraries

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

import tensorflow as tf
```

```python
data = pd.read_csv('archive/Spam Email raw text for NLP.csv')
data
```

```python
data.info()
```

### Preprocessing

```python
df = data.copy()
```

```python
# Drop FILE_NAME column
df = df.drop('FILE_NAME', axis=1)
```

```python
df
```

```python
# Split df into X and y
y = df['CATEGORY'].copy()
X = df['MESSAGE'].copy()
```

```python
X
```

```python
y
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
```

```python
X_train
```

```python
# Create tokenizer
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=30000)
```

```python
# Fit the tokenizer
tokenizer.fit_on_texts(X_train)
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
np.array(tokenizer.texts_to_sequences(X_train), dtype='object')
```

```python
seq = tokenizer.texts_to_sequences(X_train)
```

```python
np.max(list(map(lambda x: len(x), seq)))
```

```python
def get_sequences(texts, tokenizer, train=True, max_seq_length=None):
    sequences = tokenizer.texts_to_sequences(texts)
    if train == True:
        max_seq_length = np.max(list(map(lambda x: len(x), sequences)))
    sequences = tf.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=max_seq_length, padding='post')
    return sequences
```

```python
get_sequences(X_train, tokenizer, train=True).shape
```

```python
# Convert texts to sequences
X_train = get_sequences(X_train, tokenizer, train=True)
X_test = get_sequences(X_test, tokenizer, train=False, max_seq_length=X_train.shape[1])
```

### Training

```python
inputs = tf.keras.Input(shape=(14804,))

embedding = tf.keras.layers.Embedding(
    input_dim = 30000,
    output_dim = 64
)(inputs)
```

```python
inputs
```

```python
embedding
```

```python
flatten = tf.keras.layers.Flatten()(embedding)
flatten
```

```python
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(flatten)

model = tf.keras.Model(inputs=inputs, outputs=outputs)
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc')
    ]
)

print(model.summary())
```

```python
history = model.fit(
    X_train, 
    y_train,
    validation_split=0.2,
    batch_size=32,
    epochs=100,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=3,
            restore_best_weights=True
        )
    ]
)
```

### Results

```python
results = model.evaluate(X_test, y_test, verbose=0)

print("    Test Loss: {:.4}".format(results[0]))
print("Test Accuracy: {:.2f}%".format(results[1] * 100))
print("     Test AUC: {:.4}".format(results[2]))
```
