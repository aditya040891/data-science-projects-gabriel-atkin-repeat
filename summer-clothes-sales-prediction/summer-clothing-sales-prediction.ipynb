## Summer Clothing Sales Prediction

Data source: https://www.kaggle.com/datasets/jmmvutu/summer-products-and-sales-in-ecommerce-wish

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split

import tensorflow as tf
```

```python
data = pd.read_csv('archive/summer-products-with-rating-and-performance_2020-08.csv')
data
```

```python
data.info()
```

```python
# Drop unnecessary columns
columns_to_drop = ['title', 'title_orig', 'currency_buyer', 'shipping_option_name', 'urgency_text', 'merchant_title', 'merchant_name',
                  'merchant_info_subtitle', 'merchant_id', 'merchant_profile_picture', 'product_url', 'product_picture', 'product_id', 'tags',
                  'has_urgency_banner', 'theme', 'crawl_month', 'origin_country']

data = data.drop(columns_to_drop, axis=1)
```

```python
data.info()
```

### Preprocessing

```python
df = data.copy()
```

```python
# Missing values
df.isna().sum()
```

#### Encoding

```python
df
```

```python
df['product_variation_size_id'].value_counts()
```

```python
size_ordering = ['XXS', 'XS', 'S', 'M', 'L', 'XL', 'XXL']

def ordinal_encode(data, column, ordering):
    return data[column].apply(lambda x: ordering.index(x) if x in ordering else None)
```

```python
df['product_variation_size_id'] = ordinal_encode(df, 'product_variation_size_id', size_ordering)
df['product_variation_size_id']
```

```python
df['product_variation_size_id'].value_counts()
```

```python
def onehot_encode(data, column):
    dummies = pd.get_dummies(data[column], prefix=column, dtype=int)
    data = pd.concat([data, dummies], axis=1)
    data = data.drop(column, axis=1)
    return data
```

```python
df = onehot_encode(df, 'product_color')
df
```

```python
(df.dtypes == 'object').sum()
```

```python
# Missing values
df.isna().sum()[df.isna().sum()>0]
```

```python
null_columns = ['rating_five_count', 'rating_four_count', 'rating_three_count', 
                'rating_two_count', 'rating_one_count', 'product_variation_size_id']

for column in null_columns:
    df[column] = df[column].fillna(df[column].mean())
```

```python
df.isna().sum().sum()
```

#### Scaling

```python
df
```

```python
y = df['units_sold'].copy()
X = df.drop('units_sold', axis=1).copy()
```

```python
scaler = MinMaxScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X
```

```python
y
```

```python
y.value_counts()
```

#### A little more encoding

```python
encoder = LabelEncoder()

y = encoder.fit_transform(y)

y_mappings = {index: label for index, label in enumerate(encoder.classes_)}
y_mappings
```

```python
y
```

### Training

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=1)
```

```python
inputs = tf.keras.Input(shape=(124,))
x = tf.keras.layers.Dense(16, activation='relu')(inputs)
x = tf.keras.layers.Dense(16, activation='relu')(x)
outputs = tf.keras.layers.Dense(15, activation='softmax')(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

batch_size = 32
epochs = 105

history = model.fit(
    X_train,
    y_train,
    validation_split = 0.2,
    batch_size = batch_size,
    epochs = epochs
)
```

```python
plt.figure(figsize=(14, 6))

epochs_range = range(1, epochs + 1)
train_loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(epochs_range, train_loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend('upper right')
plt.show()
```

```python
np.argmin(val_loss)
```

```python
model.evaluate(X_test, y_test)
```
