## EEG Student Confusion Prediction

Given *EEG data from students during MOOC lectures*, let's try to predict whether a given student will be **confused** during a lecture.

We will use a Keras/TensorFlow neural network to make our predictions.

Data source: https://www.kaggle.com/datasets/wanghaohan/confused-eeg

### Importing Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid", color_codes=True)

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

import tensorflow as tf

from sklearn.metrics import confusion_matrix, classification_report
```

```python
eeg_df = pd.read_csv('archive/EEG_data.csv')
info_df = pd.read_csv('archive/demographic_info.csv')
```

```python
eeg_df
```

```python
info_df
```

#### Merging DataFrames 

```python
info_df.rename(columns={'subject ID': 'SubjectID'}, inplace=True)
```

```python
info_df
```

```python
data = info_df.merge(eeg_df, on='SubjectID')
data
```

```python
data.info()
```

#### Cleaning Data

```python
data = data.drop(['SubjectID', 'VideoID', 'predefinedlabel'], axis=1)
```

```python
data.columns
```

```python
data.rename(columns={' age': 'Age', ' ethnicity': 'Ethnicity', ' gender': 'Gender', 'user-definedlabeln': 'Label'}, inplace=True)
```

```python
data
```

```python
data['Label'] = data['Label'].astype(int)
```

```python
data
```

```python
print("Missing values: ", data.isna().sum().sum())
```

#### Encoding Features

```python
data['Gender'].unique()
```

```python
data['Gender'] = data['Gender'].apply(lambda x: 1 if x == 'M' else 0)
data['Gender']
```

```python
data['Ethnicity'].unique()
```

```python
ethnicity_dummies = pd.get_dummies(data['Ethnicity'], dtype=int)
data = pd.concat([data, ethnicity_dummies], axis=1)
data = data.drop('Ethnicity', axis=1)
```

```python
data
```

```python
print("Non-numeric columns: ", len(data.select_dtypes('object').columns))
```

### EDA

#### Splitting and Scaling

```python
y = data['Label'].copy()
X = data.drop('Label', axis=1).copy()
```

```python
scaler = StandardScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=123)
```

### Training

```python
inputs = tf.keras.Input(shape=(X_train.shape[1],))
x = tf.keras.layers.Dense(64, activation='relu')(inputs)
x = tf.keras.layers.Dense(64, activation='relu')(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = [
        'accuracy',
        tf.keras.metrics.AUC(name='auc')
    ]
)

epochs = 50
batch_size = 32

history = model.fit(
    X_train,
    y_train,
    validation_split = 0.2,
    batch_size = batch_size,
    epochs = epochs,
    callbacks = [
        tf.keras.callbacks.ReduceLROnPlateau()
    ]
)
```

### Results

```python
plt.figure(figsize=(8,6))
plt.plot(range(epochs), history.history['loss'], label='Training Loss')
plt.plot(range(epochs), history.history['val_loss'], label='Validation Loss')

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Over Time")
plt.legend()

plt.show()
```

```python
model.evaluate(X_test, y_test)
```

```python
y_true = np.array(y_test)

y_pred = np.squeeze(model.predict(X_test))
y_pred = np.array(y_pred >= 0.5, dtype=int)
```

```python
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(4,4))

sns.heatmap(cm, annot=True, fmt='g', vmin=0, cbar=False)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

plt.show()
```

```python
print(classification_report(y_true, y_pred))
```
