## Heart Failure Prediction

Data source: https://www.kaggle.com/datasets/andrewmvd/heart-failure-clinical-data

### Importing Libraries

```python
import numpy as np
import pandas as pd
from ydata_profiling import ProfileReport

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv('archive/heart_failure_clinical_records_dataset.csv')
data
```

```python
profile = ProfileReport(data)
```

```python
profile.to_notebook_iframe()
```

```python
y = data['DEATH_EVENT'].copy()
X = data.drop('DEATH_EVENT', axis=1)
```

```python
y
```

```python
X
```

```python
scaler = MinMaxScaler()

X = scaler.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
model_log = LogisticRegression()
model_log.fit(X_train, y_train)
y_pred_log = model_log.predict(X_test)

log_acc = model_log.score(X_test, y_test)
log_f1 = f1_score(y_test, y_pred_log)
```

```python
model_svm = SVC()
model_svm.fit(X_train, y_train)
y_pred_svm = model_svm.predict(X_test)

svm_acc = model_svm.score(X_test, y_test)
svm_f1 = f1_score(y_test, y_pred_svm)
```

```python
model_mlp = MLPClassifier(hidden_layer_sizes=(128, 128))
model_mlp.fit(X_train, y_train)
y_pred_mlp = model_mlp.predict(X_test)

mlp_acc = model_mlp.score(X_test, y_test)
mlp_f1 = f1_score(y_test, y_pred_mlp)
```

```python
print(f"Logistic Regression:\nAccuracy: {log_acc:.2f}\nF1 Score: {log_f1:.4f}\n")
print(f"Support Vector Machine:\nAccuracy: {svm_acc:.2f}\nF1 Score: {svm_f1:.4f}\n")
print(f"Multi Layer Perceptron:\nAccuracy: {mlp_acc:.2f}\nF1 Score: {mlp_f1:.4f}\n")
```
