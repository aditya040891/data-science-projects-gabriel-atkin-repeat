Data source: https://www.kaggle.com/datasets/edumagalhaes/quality-prediction-in-a-mining-process?resource=download

### Getting Started

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)

import re
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("archive/MiningProcess_Flotation_Plant_Database.csv")
data
```

```python
data.info()
```

### Preprocessing

```python
df = data.copy()
```

```python
df.isna().sum()
```

```python
for column in df.columns:
    df[column] = df[column].apply(lambda x: x.replace(',', '.'))
```

```python
df
```

### Feature Engineering

```python
df['date'] = df['date'].apply(lambda x: re.search('[0-9]*-[0-9]*', x).group(0))
df['date']
```

```python
df['Year'] = df['date'].apply(lambda x: re.search('^[^-]*', x).group(0))
df['Month'] = df['date'].apply(lambda x: re.search('[^-]*$', x).group(0))
```

```python
df = df.drop('date', axis=1)
```

```python
df
```

```python
df = df.astype(float)
```

```python
df
```

### Visualization

```python
df['Year'].unique()
```

```python
df = df.drop('Year', axis=1)
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), vmin=-1.0, vmax=1.0)
plt.show()
```

### Splitting and Scaling

```python
target = '% Silica Concentrate'
y = df[target]

X_n = df.drop([target, '% Iron Concentrate'], axis=1)
X_i = df.drop(target, axis=1)
```

```python
y
```

```python
X_n
```

```python
X_i
```

```python
scaler = StandardScaler()

X_n = scaler.fit_transform(X_n)
X_i = scaler.fit_transform(X_i)
```

```python
X_n_train, X_n_test, y_n_train, y_n_test = train_test_split(X_n, y, train_size=0.7)
X_i_train, X_i_test, y_i_train, y_i_test = train_test_split(X_i, y, train_size=0.7)
```

### Training

```python
model_n = LinearRegression()
model_i = LinearRegression()
```

```python
model_n.fit(X_n_train, y_n_train)
print("Model without iron R^2 Score: ", model_n.score(X_n_test, y_n_test))
```

```python
model_i.fit(X_i_train, y_i_train)
print("Model with iron R^2 Score: ", model_i.score(X_i_test, y_i_test))
```
